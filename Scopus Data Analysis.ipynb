## Top 20 Citations

```python
plt.figure(figsize=(15,7))
df['Citations'].sort_values(ascending=False)[:20].plot(kind='bar', color='red')
plt.xlabel('Name')
plt.ylabel('Citations')
plt.title('Top 20 Citations')
```

## Top 20 Documents

```python
plt.figure(figsize=(15,7))
df['Documents'].sort_values(ascending=False)[:20].plot(kind='bar', color='blue')
plt.xlabel('Name')
plt.ylabel('Documents')
plt.title('Top 20 Documents')
```

## Most citations in 2010

```python
plt.figure(figsize=(15,7))
df['Citations2010'].sort_values(ascending=False)[:20].plot(kind='bar', color='purple')
plt.xlabel('Name')
plt.ylabel('2010 Citations')
plt.title('Top 20 2010 Citations')
```

```python
df_topTopic = df['Topic1'][df['Topic1'] != 0]
```

```python
df_topTopic.head()
```

```python
av_topics = []
for topic_list in df_topTopic:
    topics = topic_list.split(';')
    for topic in topics:
        if topic in av_topics:
            pass
        else:
            av_topics.append(topic)
```

```python
av_topics
```

```python
topics_distribution = []
for topic in av_topics:
    topics_distribution.append(len(df_topTopic[df_topTopic.str.contains(topic)]))
```

```python
topics_distribution
```

## Top Topic Distribution

```python
plt.figure(figsize=(20,7))
plt.bar([topic[:2]+'.' for topic in av_topics], topics_distribution, color='orange');
plt.xlabel('Topics')
plt.ylabel('Entries')
plt.title('Number of entries per topic')
```
